```python
#Gráfico Histograma RPM Interativo - Plotly

trace1 = go.Histogram(x = volta1['RPM'], name = 'Volta 1', xbins = dict(start = 3000, end = 12500, size = 500))
trace2 = go.Histogram(x = volta2['RPM'], name = 'Volta 2', xbins = dict(start = 3000, end = 12500, size = 500))
trace3 = go.Histogram(x = volta3['RPM'], name = 'Volta 3', xbins = dict(start = 3000, end = 12500, size = 500))
trace4 = go.Histogram(x = volta4['RPM'], name = 'Volta 4', xbins = dict(start = 3000, end = 12500, size = 500))
trace5 = go.Histogram(x = volta5['RPM'], name = 'Volta 5', xbins = dict(start = 3000, end = 12500, size = 500))

data = [trace1, trace2, trace3, trace4, trace5]
layout = go.Layout(title = 'Histograma RPM',
                  titlefont = {'family': 'Arial',
                              'size': 22,},
                  xaxis = {'title': 'RPM'},
                  yaxis = {'title': 'Quantidade de amostras'})
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

```python
#Gráfico Histograma Velocidade Interativo - Plotly

trace1 = go.Histogram(x = volta1['GPS_Speed'], name = 'Volta 2', xbins = dict(size = 20))
trace2 = go.Histogram(x = volta2['GPS_Speed'], name = 'Volta 3', xbins = dict(start = 70, size = 20))
trace3 = go.Histogram(x = volta3['GPS_Speed'], name = 'Volta 4', xbins = dict(size = 20))
trace4 = go.Histogram(x = volta4['GPS_Speed'], name = 'Volta 5', xbins = dict(size = 20))
trace5 = go.Histogram(x = volta5['GPS_Speed'], name = 'Volta 6', xbins = dict(size = 20))

data = [trace1, trace2, trace3, trace4, trace5]
layout = go.Layout(title = 'Histograma Velocidade',
                  titlefont = {'family': 'Arial',
                              'size': 22,},
                  xaxis = {'title': 'Velocidade km/h'},
                  yaxis = {'title': 'Quantidade de amostras'})
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```
